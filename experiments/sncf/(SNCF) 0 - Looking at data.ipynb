```python
img = plt.imread('/datasets/shared_datasets/SNCF/DATASET_etat_feu/images/AWKTTZm43632.jpg')
h1,w1,_=img.shape
with open("/datasets/shared_datasets/SNCF/DATASET_etat_feu/classes.txt", 'r') as f:
    classes_raw = f.read().splitlines()
    classes_raw = [c.split(" ") for c in classes_raw if len(c) > 0]
classes = {c[0]: " ".join(c[1:]) for c in classes_raw}
print(len(classes), classes)
with open('/datasets/shared_datasets/SNCF/DATASET_etat_feu/labels/AWKTTZm43632.txt', 'r') as f:
    label = f.read()
    label = label.split(" ")
    print(label)
    box = np.array(list([float(x) for x in label[1:]]))
    box = box * np.array([w1, h1, w1, h1])
plt.imshow(img)
plt.axis('off')
# Draw rectangle with label
plt.gca().add_patch(plt.Rectangle((box[0]-box[2]/2, box[1]-box[3]/2), box[2], box[3], linewidth=2, edgecolor='r', facecolor='none'))
# plot label on the box
plt.text(box[0]-box[2]/2, box[1]-box[3]/1.5, classes[label[0]], color='white', fontsize=9, bbox=dict(facecolor='red', alpha=0.5))
```
